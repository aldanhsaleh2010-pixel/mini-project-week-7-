# import for all required library 🔧⏬

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# Load data ⌛


```python
df_EDA = pd.read_csv("/kaggle/input/datasets/dana3333/day-35555/day35_project.csv")
```

# 1️⃣first step >
## 💥 data overview - shape - Dtypes - missing value 

🔴 why we do this as first step ?
ctually, when we understand the data first, it helps guide all the analysis that follows 🧐

```python
print("1️⃣ Data types:\n", df_EDA.dtypes)
print("\n2️⃣ Shape:", df_EDA.shape)
print("\n3️⃣ Check if there are any missing values:\n", df_EDA.isna().sum())
```

# 2️⃣ second step >>
# Identify column types 🧐!

🔴 is it important ?
Yes, of course. Because the type of columns determines how we analyze the data and how we visualize it.

```python
numeric_cols = df_EDA.select_dtypes(include=["number"]).columns
print("🔢: Numeric columns\n:", numeric_cols)

categorical_cols = df_EDA.select_dtypes(include=["object", "category"]).columns
print("\n 📔: Categorical columns\n:", categorical_cols)
```

# 3️⃣ therd step >>>
# Univariate analysis 📈

🔴 why this step important ?
Because we want to understand the data shape, distribution, and identify any issues.


```python
sns.set_palette("Set2")

for col in numeric_cols[:3]:
    sns.histplot(df_EDA[col].dropna())
    plt.title(col)
    plt.show()
```

```python
sns.set_palette("pastel")

for col in categorical_cols[:2]:
    sns.countplot(x=df_EDA[col])
    plt.title(col)
    plt.show()
```

# 4️⃣ forth step >>>>
# Bivariate relationships 🔍

🔴 why this serep important ?
because it helps us understand the relationship between variables.

```python
sns.scatterplot(x=numeric_cols[0], y=numeric_cols[1], data=df_EDA)
plt.title("Relationship between two numeric variables")
plt.show()
```

```python
sns.boxplot(x=categorical_cols[0], y=numeric_cols[0], data=df_EDA)
plt.title("Numeric vs Categorical")
plt.show()
```

# 5️⃣ fifth step >>>>>
##  Correlation and Multivariate patterns!!

🔴 why this serep important ?
because it helps us understand the correlation between variables and identify strong relationships.

```python
sns.heatmap(df_EDA[numeric_cols].corr(), annot=True)
plt.title("Correlation Heatmap")
plt.show()
```
